```python
import zipfile
import io
import os
from PIL import Image
import pytesseract

# Path to the ZIP file containing folders with JPG files
zip_file_path = '/content/drive/MyDrive/Clean-Data-Resized-224x224-signature-verification.zip'

# Temporary folder to extract images
extracted_images_folder = '/content/drive/MyDrive/ExtractedContents'  # Specifying a folder to extract images

# Create a directory to store extracted images if it doesn't exist
if not os.path.exists(extracted_images_folder):
    os.makedirs(extracted_images_folder)

# Path to Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Function to extract text from an image
def extract_text_from_image(image_path):
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img, lang='eng')
    return text

# Open the ZIP file and extract images from folders
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Extract all files from the ZIP file
    zip_ref.extractall(extracted_images_folder)

# Perform OCR on extracted images within folders
for root, dirs, files in os.walk(extracted_images_folder):
    for file_name in files:
        if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(root, file_name)
            extracted_text = extract_text_from_image(image_path)
            print(f"Text extracted from {file_name}:")
            print(extracted_text)
```
